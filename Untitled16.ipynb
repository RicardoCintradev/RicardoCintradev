```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Carregar o dataset (ajustado para o caminho do Colab)
df = pd.read_csv('cyberfeddefender_dataset.csv')

print("--- Resumo do Dataset ---")
print(f"Total de registros: {len(df)}")
print(f"Distribuição de Ataques:\n{df['Attack_Type'].value_counts()}")

# Gráfico 1: Distribuição de Ataques
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Attack_Type', palette='viridis')
plt.title('Distribuição dos Tipos de Ataque')
plt.show()

# Gráfico 2: Protocolos por Ataque
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Attack_Type', hue='Protocol')
plt.title('Protocolos Utilizados em cada Tipo de Ataque')
plt.legend(title='Protocolo')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Preparação dos dados
features = df.drop(['Timestamp', 'Source_IP', 'Destination_IP', 'Attack_Type', 'Label'], axis=1)
target = df['Attack_Type']

# Codificar Protocolo e Flags
le = LabelEncoder()
features['Protocol'] = le.fit_transform(features['Protocol'])
features['Flags'] = le.fit_transform(features['Flags'])

# Dividir Treino e Teste
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Treinar Modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Mostrar Resultados
y_pred = model.predict(X_test)
print("\n--- Relatório de Classificação ---")
print(classification_report(y_test, y_pred))

# Matriz de Confusão
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.title('Matriz de Confusão - Detecção de Ataques')
plt.show()
```

```python
import random

class CyberSecurityEnv:
    def __init__(self, dataframe):
        self.df = dataframe
        self.action_space = ["Allow", "Block_IP", "Rate_Limit", "Isolate"]

    def step(self, idx, action_idx):
        row = self.df.iloc[idx]
        is_attack = row['Label'] == 1
        action = self.action_space[action_idx]

        # Lógica de Recompensa (O Agente de RL quer maximizar isso)
        if is_attack:
            reward = 10 if action != "Allow" else -10
        else:
            reward = 5 if action == "Allow" else -5

        return reward, action, row['Attack_Type']

# Simulação de 15 fluxos de rede
env = CyberSecurityEnv(df)
print(f"{'Passo':<6} | {'Tipo Real':<15} | {'Ação da IA':<12} | {'Resultado'}")
print("-" * 55)

for i in range(15):
    action_idx = random.randint(0, 3) # Simula a escolha do agente
    reward, action, attack = env.step(i, action_idx)
    status = "✅ ACERTO" if reward > 0 else "❌ ERRO"
    print(f"{i+1:<6} | {attack:<15} | {action:<12} | {status} (Pts: {reward})")
```
